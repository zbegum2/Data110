```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(titanic['fare'], bins=30, color='red')
plt.title('Fare Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(titanic['fare'], color='red', fill=True)
plt.title('KDE of Fare')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.show()

```

It appears that there is the most amount of counts of fares under the price of 100 pounds. The outlier of 500 pounds is the lowest of all of the prices. Which means that we can infer that there were much more lower-class individuals because of the price paid for certain fares.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(data=titanic, x='fare', hue='survived', fill=True, common_norm=False, palette={0: 'green', 1: 'red'})

plt.legend(title='Survived', labels=['No', 'Yes'])

plt.title('Fare by Survival Status')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.show()
```

Using this specific plot, it is not quite a drastic different between fare and survival rates, just fromt he kde plot. We would need a better comparison because looking at this graph it seems the lower fares paid were more likely to survive. But also, we can see that the width of the non-survivors plot is much thicker than the survivors' plot, under 100 pounds. So this could be something to note does that there more death between 0-100 pounds paid rather than survivors.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x='fare', hue='pclass', multiple='stack', bins=30, kde=False)

plt.title('Fare by Pclass')
plt.xlabel('Fare')
plt.ylabel('Count')

plt.show()
```

Looking at this histogram, it is more clear that the 1st class of passengers, the ones with more wealth, bought the more expensive tickets.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x='survived', hue='pclass', palette={1: 'green', 2: 'blue', 3: 'brown'})

plt.title('Survival by Pclass')
plt.xlabel('Survival 0=No, 1=Yes')
plt.ylabel('Count')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.kdeplot(data=titanic, x='survived', hue='pclass', fill=True, common_norm=False, palette={1: 'skyblue', 2: 'lightgreen', 3: 'orange'})

plt.title('KDE Plot of Survival by Pclass')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Density')

plt.show()
```

On the left side of the kde and histogram plots, we have the non-survivors' pclasses and it seems that the passengers that were 1st class have the least amount of deaths and the passengers that were 3rd class, the lowest, had the highest amount of deaths. On the right side we can see the survivors' pclasses and the highest survivors were 1st class. Seeing that passengers who are more wealthy or bought better tickets have a higher rate of survival makes it easy to infer that their social status must have been higher and mroe important at the time. So when in an emergency, priority was given to the higher class, since they paid more for their tickets. Even though at the end of the day, we are all human, social classes will always have an importance when there is a large amount of people gathered in one area.
