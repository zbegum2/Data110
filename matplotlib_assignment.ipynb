```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 2*x + 1

x = np.linspace(-2, 4, 100)
y = f(x)

plt.plot(x, y, label=r'$f(x) = x^2 - 2x + 1$')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the quadratic function $f(x) = x^2 - 2x + 1$')

plt.grid(True)
plt.legend()


plt.show()


```

```python

import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return x**3

def g(x):
    return np.exp(x)

def h(x):
    return np.log(x)


x = np.linspace(-2, 2, 100)
x_positive = np.linspace(0.1, 2, 100)


y_f = f(x)
y_g = g(x)
y_h = h(x_positive)


plt.plot(x, y_f, label=r'$f(x) = x^3$', color='blue')
plt.plot(x, y_g, label=r'$g(x) = e^x$', color='green')
plt.plot(x_positive, y_h, label=r'$h(x) = \log(x)$', color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of functions $f(x) = x^3$, $g(x) = e^x$, and $h(x) = \log(x)$')


plt.legend()


plt.grid(True)

plt.show()
```

```python

def f(x):
    return x**2

def g(x):
    return x**3

def h(x):
    return x**4

def i(x):
    return x**5

x = np.linspace(-2, 2, 100)

y_f = f(x)
y_g = g(x)
y_h = h(x)
y_i = i(x)


plt.plot(x, y_f, label=r'$f(x) = x^2$', color='red', linestyle='-')
plt.plot(x, y_g, label=r'$g(x) = x^3$', color='#FF5733', linestyle='--')
plt.plot(x, y_h, label=r'$h(x) = x^4$', color=(0.2, 0.6, 0.4), linestyle='-.')
plt.plot(x, y_i, label=r'$i(x) = x^5$', color='0.5', linestyle=':')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of functions $f(x) = x^2$, $g(x) = x^3$, $h(x) = x^4$, and $i(x) = x^5$ with different color methods')


plt.legend()

plt.grid(True)

plt.show()
```

```python

def f(x):
    return 1 / (x**2 + 1)

x = np.linspace(-10, 10, 400)

y = f(x)

plt.figure(figsize=(12, 6))

# First plot
plt.subplot(1, 2, 1)
plt.plot(x, y, label=r'$f(x) = \frac{1}{x^2 + 1}$', color='blue')
plt.xlim(-10, 10)
plt.ylim(0, 1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of $f(x) = \frac{1}{x^2 + 1}$ with Custom Limits')
plt.grid(True)
plt.legend()

# Second plot with reversed x&y axes
plt.subplot(1, 2, 2)
plt.plot(y, x, label=r'$f(x) = \frac{1}{x^2 + 1}$', color='red')
plt.xlim(0, 1)
plt.ylim(-10, 10)
plt.xlabel('f(x)')
plt.ylabel('x')
plt.title('Reversed Axes Plot of $f(x) = \frac{1}{x^2 + 1}$')
plt.grid(True)
plt.legend()

# Show the plots with tight and equal axis settings
plt.figure(figsize=(12, 6))

# Tight axis
plt.subplot(1, 2, 1)
plt.plot(x, y, color='blue')
plt.axis('tight')
plt.title('Tight Axis Setting')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Equal axis
plt.subplot(1, 2, 2)
plt.plot(x, y, color='green')
plt.axis('equal')
plt.title('Equal Axis Setting')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```
