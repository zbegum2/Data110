```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/us_contagious_diseases.csv')

df.head()

```

```python
from matplotlib import pyplot as plt
import seaborn as sns

def _plot_series(series, series_name, series_index=0):
    palette = list(sns.palettes.mpl_palette('Dark2'))
    xs = series['year']
    ys = series['count']

    # Use plt.scatter for a scatter plot with smaller dots
    plt.scatter(xs, ys, label=series_name, color=palette[series_index % len(palette)], s=30)  # Adjusted size to 30

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('year', ascending=True)

for i, (series_name, series) in enumerate(df_sorted.groupby('disease')):
    _plot_series(series, series_name, i)

fig.legend(title='Disease', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Year')
plt.ylabel('Count')

plt.title('Disease Counts Over Time')  # Optional: add a title for clarity
# Removed the grid
plt.show()

```

```python
from matplotlib import pyplot as plt
import seaborn as sns

def _plot_series(series, series_index=0):
    palette = list(sns.palettes.mpl_palette('Dark2'))
    ys = series['state']  # Use 'state' for y-axis
    xs = series['count']  # Use 'count' for x-axis

    # Create a horizontal bar plot
    plt.barh(ys, xs, color=palette[series_index % len(palette)], alpha=0.7)

# Increase the figure height for better readability
fig, ax = plt.subplots(figsize=(10, 15), layout='constrained')  # Adjusted height to 15
df_sorted = df.sort_values('state')  # Sort by 'state'

# Grouping by 'state' and including 'disease' data
for i, (state, series) in enumerate(df_sorted.groupby(['state', 'disease'])):
    _plot_series(series, i)

sns.despine(fig=fig, ax=ax)
plt.xlabel('Count')  # Updated x-axis label to 'Count'
plt.ylabel('State')  # Updated y-axis label to 'State'

plt.title('Disease Counts by State')  # Optional: add a title for clarity

# Remove the legend by not including plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

# Filter out specific diseases
diseases_to_remove = ['Mumps', 'Pertussis', 'Smallpox', 'Polio']
df_filtered = df[~df['disease'].isin(diseases_to_remove)]  # Keep only the diseases not in the list

def _plot_series(series, series_index=0):
    xs = series['population']  # Horizontal bar width corresponds to 'population'
    ys = series['state']  # Vertical axis corresponds to 'state'

    # Create a horizontal bar plot with a single color
    plt.barh(ys, xs, color='green')  # Changed to a single color for all bars

# Increase the figure height for better readability
fig, ax = plt.subplots(figsize=(10, 15), layout='constrained')  # Adjusted height to 15 for better spacing
df_sorted = df_filtered.sort_values(['year', 'state'], ascending=[True, True])  # Sort by 'year' and 'state'

# Grouping by the 'state' column
for i, (state, series) in enumerate(df_sorted.groupby('state')):  # Loop through grouped states
    _plot_series(series, i)  # Pass the data to _plot_series

# Apply formatting to x-axis labels for large numbers
ax.xaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{int(x):,}'))

sns.despine(fig=fig, ax=ax)
plt.xlabel('Population')  # Updated x-axis label to 'Population'
plt.ylabel('State')  # Updated y-axis label to 'State'
plt.title('Population by State Over Time')  # Optional: add a title for clarity
plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

# Create a new DataFrame without the 'disease' column
df_filtered = df.drop(columns=['disease'])

def _plot_series(series, series_index=0):
    palette = ['green']  # Set a single color for all lines
    xs = series['population']
    ys = series['year']  # Changed from 'count' to 'population'

    # Plot as a line instead of bars
    plt.plot(xs, ys, color=palette[0], label=f'Series {series_index + 1}')  # Line color set to green

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df_filtered.sort_values('year', ascending=True)  # Sort without the 'disease' column

# Grouping by the 'year' column
for i, (year, series) in enumerate(df_sorted.groupby('year')):  # Loop through grouped years
    _plot_series(series, i)  # Pass the data to _plot_series

# Change legend to show only 'Population'
plt.legend(['Population'], title='Population', loc='upper left')  # Static legend
sns.despine(fig=fig, ax=ax)
plt.xlabel('Population')
plt.ylabel('Year')  # Updated y-axis label to 'Population'

plt.title('Population Over Time')  # Optional: add a title for clarity
plt.show()
```
