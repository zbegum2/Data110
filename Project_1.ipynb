```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/us_contagious_diseases.csv')

df.head()

```

```python
from matplotlib import pyplot as plt
import seaborn as sns

def _plot_series(series, series_name, series_index=0):
    palette = list(sns.palettes.mpl_palette('Dark2'))
    xs = series['year']
    ys = series['count']

    plt.scatter(xs, ys, label=series_name, color=palette[series_index % len(palette)], s=30)

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('year', ascending=True)

for i, (series_name, series) in enumerate(df_sorted.groupby('disease')):
    _plot_series(series, series_name, i)

fig.legend(title='Disease', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Year')
plt.ylabel('Count')

plt.title('Disease Counts Over Time')

plt.show()

```

This scatterplot was made to show the exact oulier which is Measles around 1940 where is higher than any other disease. This is important to glance at the way it is decreasing also, all because of the vaccine being created.
