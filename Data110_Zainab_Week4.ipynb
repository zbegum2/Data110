```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")
```

```python
x = ['Assault', 'Burglary', 'Larceny', 'Car Theft', 'Murder', 'Rape', 'Robbery']
y = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]

plt.bar(x, y, label='Amount of Crimes')
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.title('Crimes in 2022-2023')
plt.legend()
plt.show()
```

```python
x = ['Assault', 'Burglary', 'Larceny', 'Car Theft', 'Murder', 'Rape', 'Robbery']
y = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]

plt.ylabel('Offenses per 100,000 People')
plt.xlabel('Offense')
plt.bar(x, y, label='Amount of Crimes', color=['red', 'green', 'orange', 'blue', 'purple', 'grey', 'pink'])
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.title('Crimes in 2022-2023')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = ['Assault', 'Burglary', 'Larceny', 'Car Theft', 'Murder', 'Rape', 'Robbery']
y = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]

plt.xlabel('Change Rate')
plt.ylabel('Offense')
plt.barh(x, y, label='Amount of Crimes', color=['teal', 'sienna', 'silver', 'plum', 'purple', 'coral', 'brown'])
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.title('Crimes in 2022-2023')
plt.legend()
plt.grid(True)
plt.show()
```

Why does this bar graph make more sense than Task 1's bar graph?

On the first graph it seems as though Larceny stands out and has the most crimes therefore it might be the most common, but in the second graph, you can see that Car Theft drastically increased and Larceny actually **decreased**. In this case it seems more accurate and resourceful in using the second graph to make an overall statement or view of the crime rates. Seeing the change over time is more important because this can help officials make the proper changes and shift their focus onto the crimes that have been increasing.
