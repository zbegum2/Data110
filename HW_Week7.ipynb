```python
import seaborn as sns
import pandas as pd

iris = sns.load_dataset('iris')

df = iris

iris.head()
```

```python
iris.describe()
```

Pair Plot:

```python
sns.pairplot(iris, hue= 'species')
plt.suptitle("Combination of Features", y=1.02, fontsize=16)
```

Analysis: Using a pairplot is very helpful when comparing the different lengths and widths of the species. The scatterplots help to see a wider range of differences between two different features and how their combinations affect rankings. On the bottom left of the plot, we can see that Virginica is overall the highest in both sepal_length and petal_width which can be really interesting to note down when comparies these species. The KDE graphs are also a nice way to look at one overall feature to narrow down a ranking among the species.

Bar Plot:

```python
plt.bar(iris['species'],iris['sepal_length'])
plt.title('Species Sepal Lengths')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
```

Analysis: The bar plot is the easiest plot to read because it clearly shows what you need to know from the comparison of one feature. In this case, Virginica is clearly has the longest sepal length from all the species. Using this bar plot, we could see how each feature differentiates and see if Virginica also has the longest sepal width or petal length.

Box Plot:

```python
sns.boxplot(x='species', y='petal_width', data=df, color="skyblue")
plt.title('Petal Width by Species')
```

Analysis: Using a regualr boxplot is so useful when seeing outliers because it shows the variability of all the species and makes it easy for the reader to make an assumption or just plainly read the graph. In this plot, Virginica is the clear outlier or the highest on in petal width, we can see the median of virginica's data is higher than both setosa and versicolor. This is exeptionally helpful because we can see that versicolor has an outlier at 1.7 petal width, which is close to virginica's median, this tells us that virginica overall has the longest petal width. This would be really fun to use when comparing way more varieties of the same thing, such as different flower stem types.

Grouped Bar Plot:

```python
sns.boxplot(x="petal_length", y="sepal_length", hue="species", data=iris)
plt.xticks(rotation=90)
plt.title('Petal x Sepal Lengths')
plt.show()
```

Analysis: Although the graph looks a bit messy, we can clearly see that Virginica has a higher range of data than the other two species. It looks almost like stairs because of the data gradually moving upwards to show the different combinations of the lengths.

Violin Plot

```python
sns.violinplot(x='species', y='sepal_length', data=df, inner=None, color="orange")
plt.title('Sepal Length')
```

Analysis: The violin plot is very cool to look at and knowing the different properties it has makes it easier to read. The setosa having the widest graph at 5 means that it has a higher density of data in that specific length. For the other two species, their graphs were not as wide and have a more even graph.

Conclusion: The best overall plot for this dataset was the Pair plot because it shows exactly what we would need for an overall analysis. We can see each different species with its' highest and lowest measurements of features, and it is the easiest plot to compare. The worst in my opinion is the grouped boxplot because it does not seem to be easy to read and can confuse a lot of people with what data we actually need to focus on. One of my favorites is the box plot because you can put as much data as you want on it, and you will still be able to read it properly because you already know what outliers to look for, and the variation is very obvious.
